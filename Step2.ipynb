```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt

df = pd.read_csv("train.csv")
```

1.Distribution of Age

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'].dropna(),kde=True,bins = 30 , color = "skyblue")
plt.title('Distrubution of passenger age')
plt.xlabel("Age")
plt.ylabel("count")
plt.show()
```

 2. Fare Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Fare'], kde=True, bins=40, color='salmon')
plt.title("Distribution of Fare")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

3. Survival Countplot



```python
sns.countplot(x='Survived', data=df, palette='Set2')
plt.title("Survival Count (0 = No, 1 = Yes)")
plt.show()
```

```python
sns.countplot(x='Survived', data=df, hue='Survived', palette='Set2', legend=False)
plt.title("Survival Count (0 = No, 1 = Yes)")
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df, palette='pastel')
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df, palette='muted')
plt.title("Survival by Passenger Class")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df, palette='Set3')
plt.title("Age Distribution by Survival")
plt.show()
```

Heatmap for Missing Data

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title("Missing Data Heatmap")
plt.show()
```
